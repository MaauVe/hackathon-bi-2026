# Limpieza

```python
# 1. Filtrar solo viviendas activas (donde 'deleted_at' es Nulo)
df_viv_activas = df_Viviendas[df_Viviendas['deleted_at'].isnull()].copy()

# 2. Rellenar posibles nulos en saldos con 0
df_viv_activas['saldo_deudor'] = df_viv_activas['saldo_deudor'].fillna(0)
df_viv_activas['saldo_favor'] = df_viv_activas['saldo_favor'].fillna(0)

# 3. Cruzar Viviendas con Fraccionamientos usando el prefijo de la clave
df_Fraccionamientos['Clave'] = df_Fraccionamientos['Clave'].astype(str).str.strip()
claves = df_Fraccionamientos['Clave'].sort_values(key=lambda x: x.str.len(), ascending=False).tolist()

def extract_clave(val):
    val_str = str(val).strip().upper()
    for c in claves:
        if val_str.startswith(c.upper() + '-') or val_str == c.upper():
            return c.upper()
    return val_str.split('-')[0] if '-' in val_str else val_str

df_viv_activas['Clave_match'] = df_viv_activas['clave'].apply(extract_clave)
df_Fraccionamientos['Clave_upper'] = df_Fraccionamientos['Clave'].str.upper()
df_master_morosidad = pd.merge(df_viv_activas, df_Fraccionamientos[['Clave_upper', 'Fraccionamiento']],
                               left_on='Clave_match',
                               right_on='Clave_upper',
                               how='left')

# Rellenar los fraccionamientos que no hicieron match con 'ZONA ' + prefijo
df_master_morosidad['Fraccionamiento'] = df_master_morosidad['Fraccionamiento'].fillna('ZONA ' + df_master_morosidad['Clave_match'])

# 4. Seleccionar columnas y calcular morosidad real
df_master_morosidad['deuda_neta'] = df_master_morosidad['saldo_deudor'] - df_master_morosidad['saldo_favor']
df_master_morosidad['estatus_pago'] = np.where(df_master_morosidad['deuda_neta'] > 0, 'Deudor', 'Al Corriente')

columnas_morosidad = ['id', 'clave', 'Fraccionamiento', 'saldo_deudor', 'saldo_favor', 'deuda_neta', 'estatus_pago', 'created_at']
df_master_morosidad = df_master_morosidad[columnas_morosidad]
df_master_morosidad['Fraccionamiento'] = df_master_morosidad['Fraccionamiento'].str.strip().str.upper()
print('Viviendas procesadas:', len(df_master_morosidad))
print('Viviendas con Fraccionamiento nulo:', df_master_morosidad['Fraccionamiento'].isnull().sum())
print(df_master_morosidad.head())
```

```python
# Limpieza de Consumo de Agua
df_Consumo_de_Agua['Fpago'] = pd.to_datetime(df_Consumo_de_Agua['Fpago'], errors='coerce')
df_Consumo_de_Agua['tipo_servicio'] = 'OTROS'
df_Consumo_de_Agua.loc[df_Consumo_de_Agua['descripcion'].str.contains('MEDIDO', na=False, case=False), 'tipo_servicio'] = 'MEDIDO'
df_Consumo_de_Agua.loc[df_Consumo_de_Agua['descripcion'].str.contains('FIJA', na=False, case=False), 'tipo_servicio'] = 'CUOTA FIJA'

# Vincular Consumo con Lotes y Fraccionamiento
df_consumo_mapeado = pd.merge(df_Consumo_de_Agua, df_Lotes[['id_lote', 'Fraccionamiento']], left_on='lote', right_on='id_lote', how='left')
df_consumo_mapeado = pd.merge(df_consumo_mapeado, df_Fraccionamientos[['Id_fraccionamiento', 'Fraccionamiento']], left_on='Fraccionamiento', right_on='Id_fraccionamiento', how='left', suffixes=('_lote', '_real'))
df_consumo_mapeado['Fraccionamiento'] = df_consumo_mapeado['Fraccionamiento_real'].fillna('DESCONOCIDO').str.upper()
cols_consumo = ['Id_detalleordencompra', 'Fpago', 'TOTAL', 'cantidad', 'tipo_servicio', 'Fraccionamiento', 'lote']
df_consumo_mapeado = df_consumo_mapeado[[c for c in cols_consumo if c in df_consumo_mapeado.columns]]
print('Registros de consumo procesados:', len(df_consumo_mapeado))
print(df_consumo_mapeado.head())
```

```python
# Limpieza de Finanzas
df_master_finanzas = df_Reporte_Ingresos_Egresos.dropna(subset=['Fecha', 'concepto']).copy()
df_master_finanzas['entra'] = df_master_finanzas['entra'].fillna(0)
df_master_finanzas['sale'] = df_master_finanzas['sale'].fillna(0)
df_master_finanzas['Fecha'] = pd.to_datetime(df_master_finanzas['Fecha'], errors='coerce')

df_master_finanzas['categoria'] = 'OTROS'
df_master_finanzas.loc[df_master_finanzas['concepto'].str.contains('AGUA', na=False, case=False), 'categoria'] = 'PAGOS AGUA'
df_master_finanzas.loc[df_master_finanzas['concepto'].str.contains('NOMINA|SUELDO', na=False, case=False), 'categoria'] = 'GASTOS OPERATIVOS'

cols_a_mantener = ['Fecha', 'concepto', 'categoria', 'entra', 'sale', 'Clasificacion']
df_master_finanzas = df_master_finanzas[cols_a_mantener]
print('Registros financieros procesados:', len(df_master_finanzas))
print(df_master_finanzas.head())
```

```python
# Formatear fechas a DD/MM/YYYY antes de exportar
df_master_morosidad['created_at'] = pd.to_datetime(df_master_morosidad['created_at'], errors='coerce').dt.strftime('%d/%m/%Y')
df_consumo_mapeado['Fpago'] = pd.to_datetime(df_consumo_mapeado['Fpago'], errors='coerce').dt.strftime('%d/%m/%Y')
df_master_finanzas['Fecha'] = pd.to_datetime(df_master_finanzas['Fecha'], errors='coerce').dt.strftime('%d/%m/%Y')

# Guardar los DataFrames listos
import os
os.makedirs('../data/processed', exist_ok=True)
df_master_morosidad.to_csv('../data/processed/BI_Morosidad_Viviendas_Clean.csv', index=False, encoding='utf-8')
df_consumo_mapeado.to_csv('../data/processed/BI_Consumo_Agua_Clean.csv', index=False, encoding='utf-8')
df_master_finanzas.to_csv('../data/processed/BI_Flujo_Finanzas_Clean.csv', index=False, encoding='utf-8')
print('¡Archivos guardados en data/processed/ exitosamente con fechas en formato DD/MM/YYYY!')
```
