# Limpieza

```python
# 1. Filtrar solo viviendas activas (donde 'deleted_at' es Nulo)
df_viv_activas = df_Viviendas[df_Viviendas['deleted_at'].isnull()].copy()

# 2. Rellenar posibles nulos en saldos con 0
df_viv_activas['saldo_deudor'] = df_viv_activas['saldo_deudor'].fillna(0)
df_viv_activas['saldo_favor'] = df_viv_activas['saldo_favor'].fillna(0)

# 3. Cruzar (Merge) Viviendas con Fraccionamientos para tener el nombre del fraccionamiento
df_master_morosidad = pd.merge(df_viv_activas, df_Fraccionamientos, 
                               left_on='fraccionamiento_id', 
                               right_on='Id_fraccionamiento', 
                               how='left')

# 4. Seleccionar y renombrar solo las columnas que dan valor al negocio
columnas_morosidad = [
    'id', 'clave', 'Fraccionamiento', 'saldo_deudor', 'saldo_favor', 'created_at'
]
df_master_morosidad = df_master_morosidad[columnas_morosidad]

# Limpiar espacios en los nombres de fraccionamientos para evitar duplicados en gráficas
df_master_morosidad['Fraccionamiento'] = df_master_morosidad['Fraccionamiento'].str.strip().str.upper()

# Mostrar resultado
print(df_master_morosidad.head())
```

```python
# 1. Eliminar columnas que no aportan valor y tienen exceso de nulos
cols_a_eliminar = ['ccostostx', 'Ffact', 'ImpNombre', 'CC', 'ccostos', 'Referencia', 'Clasificacion', 'SSCanterior', 'OP']
df_master_finanzas = df_Reporte_Ingresos_Egresos.drop(columns=cols_a_eliminar)

# 2. Rellenar valores de entrada y salida vacíos con 0
df_master_finanzas['entra'] = df_master_finanzas['entra'].fillna(0)
df_master_finanzas['sale'] = df_master_finanzas['sale'].fillna(0)

# 3. Convertir la Fecha a un formato estándar de tiempo (DateTime)
df_master_finanzas['Fecha'] = pd.to_datetime(df_master_finanzas['Fecha'], errors='coerce')

# 4. Eliminar las filas (que son muy pocas) que no tienen fecha o concepto
df_master_finanzas = df_master_finanzas.dropna(subset=['Fecha', 'concepto'])

# Mostrar resultado
print(df_master_finanzas.head())
```

```python
# Guardar los DataFrames como CSV sin el índice
df_master_morosidad.to_csv("../data/processed/BI_Morosidad_Viviendas.csv", index=False, encoding="utf-8")
df_master_finanzas.to_csv("../data/processed/BI_Flujo_Finanzas.csv", index=False, encoding="utf-8")

print("¡Archivos exportados exitosamente! Listos para el Dashboard.")
```
